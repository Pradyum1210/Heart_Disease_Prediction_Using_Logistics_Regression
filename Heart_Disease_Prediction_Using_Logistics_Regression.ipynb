| Column       | Meaning                 | Simple Explanation                            |
| ------------ | ----------------------- | --------------------------------------------- |
| **age**      | Age                     | Person’s age in years                         |
| **sex**      | Sex                     | `1 = Male`, `0 = Female`                      |
| **cp**       | Chest Pain Type         | Type of chest pain (0–3)                      |
| **trestbps** | Resting Blood Pressure  | BP when person is at rest (mm Hg)             |
| **chol**     | Serum Cholesterol       | Cholesterol level (mg/dl)                     |
| **fbs**      | Fasting Blood Sugar     | `1 = >120 mg/dl`, `0 = ≤120 mg/dl`            |
| **restecg**  | Resting ECG             | ECG results (0,1,2)                           |
| **thalach**  | Max Heart Rate          | Highest heart rate achieved                   |
| **exang**    | Exercise Induced Angina | `1 = Yes`, `0 = No`                           |
| **oldpeak**  | ST Depression           | Heart stress compared to rest                 |
| **slope**    | Slope of ST Segment     | Shape of heart signal during exercise         |
| **ca**       | Major Vessels           | Number of major vessels (0–3)                 |
| **thal**     | Thalassemia             | Blood disorder type (normal/fixed/reversible) |
| **target**   | Heart Disease           | `1 = Disease`, `0 = No Disease`               |


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/heart_disease_data.csv')
```

```python
x = df.iloc[:,0:-1]
y = df['target']
```

```python
x
```

```python
y
```

```python
df['target'].value_counts()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42 , stratify = y)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
x_train_prediction = model.predict(x_train)
```

```python
training_accuracy = accuracy_score(x_train_prediction , y_train)
```

```python
x_test_accuracy = model.predict(x_test)
```

```python
testing_accuracy = accuracy_score(x_test_accuracy , y_test)
```

```python
print(f'Training Accuracy: {training_accuracy}')
print(f'Testing Accuracy: {testing_accuracy}')
```
